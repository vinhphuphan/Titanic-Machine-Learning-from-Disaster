## 1. Introduction

The sinking of the RMS Titanic is one of the most infamous shipwrecks in history. On April 15, 1912, during her maiden voyage, the Titanic sank after colliding with an iceberg, killing 1502 out of 2224 passengers and crew. This sensational tragedy shocked the international community and led to better safety regulations for ships.

One of the reasons that the shipwreck led to such loss of life was that there were not enough lifeboats for the passengers and crew. Although there was some element of luck involved in surviving the sinking, some groups of people were more likely to survive than others, such as women, children, and the upper-class.

In this portfolio, my ultimate goal is to complete the analysis of what sorts of people were likely to survive. In particular, I will take the advantage of knowledge about data science and machine learning to apply the tools of machine learning to predict which passengers survived the tragedy.

## 4. Data Preprocessing

In this step, I will perform the following preprocessing steps:
- Drop the unnecessary columns
- Drop the missing values
- Data Transformation
- Feature Engineering
- Feature Encoding

### 4.1 Drop all unnecessary columns for model and keep the dataset simple

```python
unnecess_cols = ['Ticket', 'Cabin']
titanic_df.drop(unnecess_cols,axis = 1, inplace = True)

titanic_df.head(10)
```

### 4.2 Clean missing values

#### Count the missing value

```python
titanic_df.isna().sum().sort_values(ascending = False)
```

First, we will handle the missing values of `Embarked` feature. To fill the missing values, I will find the most frequent value of `Embarked` and fill missing values with this value.

#### Find the most frequent value of Embarked

```python
mode = titanic_df['Embarked'].mode()[0]
mode
```

#### Fill the missing values with mode

```python
titanic_df['Embarked'].fillna(mode, inplace = True)
```

#### Check null value again

```python
titanic_df.isna().sum().sort_values(ascending = False)
```

We have cleaned the missing values of `Embarked` feature. Now, we need to fill the missing values in the `Age` column. The goal is to use the most correlated features with `Age` to predict the age. First, it is necessary to convert `Sex` and `Embarked` into numerical values.

```python
# Convert Sex into numerical feature where 0 = male and 1 = Female

titanic_df['Sex'] = titanic_df['Sex'].map({'male': 0, 'female': 1})
# Convert Emabarked into numerical feature where 0 = C, 1 = S and 2 = Q

titanic_df['Embarked'] = titanic_df['Embarked'].map({'C': 0, 'S': 1, 'Q' : 2})

# Display first 10 rows
titanic_df.head(10)
```

Now, we need to find the relation between `Age` and other features to fill the missing values of `Age` by predicting them.

#### Fill the missing values of Age

```python
# Change the datatype of all columns into numerical:

for i in input_features:
    titanic_df[i] = titanic_df[i].astype('Int64')

# the correlation between Age column and other columns

corr = titanic_df.drop(['Name'], axis = 1).corr()
corr
```

`Age` is negatively correlated with `SibSp`, `Parch` and `Pclass`, so, the method is that I will loop each missing ages in the `Age` column to locate the rows that have the same SibSp, Parch and Pclass values and fill the missing value with the median of these rows. Beside that, if the rows are not found, I simply fill the missing values with median of entire Age column. 

```python
# Get the indecies of null values of Age column

age_null_index = list(titanic_df[titanic_df['Age'].isnull()].index) 

for index in age_null_index:
    predict_age = titanic_df.loc[(titanic_df['SibSp'] == titanic_df.iloc[index]['SibSp'])
                                 & (titanic_df['Parch'] == titanic_df.iloc[index]['Parch'])
                                 & (titanic_df['Pclass'] == titanic_df.iloc[index]["Pclass"]), ['Age']].median()
    titanic_df['Age'].iloc[index] = predict_age
```

```python
# Fill all other missing values with median:

median_age = titanic_df['Age'].median()
titanic_df['Age'].fillna(median_age, inplace = True)
```

```python
# Check the missing values of Age

titanic_df['Age'].isna().sum()
```

#### 4.4 Feature Engineering

In this process, I will create new features base on existing features better represent the underlying problem to the predictive models resulting in improved model accuracy on unseen data. More specifically, I will create 2 new features : 
- `isAlone` 
- `Title`.

#### 4.4.1 `isAlone`

This feature will be the combination of `SibSp` and `Parch` because two existing features are aimed to show that whether the passenger on the Titanic has the family members with them or not. 

```python
# Calculate the family size

titanic_df['family_size'] = titanic_df['SibSp'] + titanic_df['Parch'] + 1
titanic_df[['SibSp', 'Parch', 'family_size']].head(10)
```

```python
# Create isAlone feature

titanic_df['isAlone'] = 0
titanic_df.loc[titanic_df['family_size'] ==  1, 'isAlone'] = 1
```

```python
# Drop SibSp, Parch, family_size features

titanic_df.drop(['SibSp', 'Parch', 'family_size'], axis = 1, inplace = True)
titanic_df.head(10)
```

Now, we have `isAlone` feature which represents whether this guest has family members on the boat or not.

#### 4.4.2 Title

When we look at the `Name` column, there is the name title in the middle of the name. These titles can be 'Mr', 'Dr', 'Master', or 'Mrs', and this title may partially show the level of the passenger in the society. Therefore, it becomes quite interesting to make this a new feature to predict whether the title of a person affects their possibility of survival after the disaster.

```python
# Get the title from `Name`

titanic_df['title'] = [name.split(',')[1].split('.')[0].strip() for name in titanic_df['Name']]
titanic_df[['Name','title']].head(10)
```

```python
# Value counts of the title

titanic_df['title'].value_counts()
```

I will get rid of 'Mr', 'Miss', 'Mrs' because we already have had `Sex` column to represent the gender. In addition, I aim to focus on 'Master', 'Dr'. Beside that, other titles will be put in one group, namely 'rare_title'.

```python
# Simplify title

titanic_df['title'] = titanic_df['title'].replace(['Rev', 'Col', 'Major', 'Lady', 'Jonkheer', 'Don', 'Capt', 'the Countess',
                                             'Sir', 'Dona'], 'Rare')
titanic_df['title'] = titanic_df['title'].replace(['Mlle', 'Ms'], 'Miss')
titanic_df['title'] = titanic_df['title'].replace('Mme', 'Mrs')
```

```python
sns.countplot(titanic_df['title'])
```

### 4.5 Feature Encoding

Machine learning models require all input and output variables to be numeric. Therefore, we need to encode all of our categorical data before we can fit the models to our data.

Previously, we have encoded the `Sex` and `Embark` column such that 0 = male and 1 = female. We need to repeat this process for Title. In addition, I will also transform `Age`, `Fare` into an ordinal variable rather than a continuous variable.

### `Title`

```python
# Drop the Name column
titanic_df.drop(['Name'], axis = 1, inplace = True)

# Encode the title column and only keep the Dr and Master title 

titanic_df = pd.get_dummies(titanic_df, columns = ['title'])
titanic_df.drop(['title_Mr', 'title_Mrs', 'title_Miss', 'title_Rare'], axis = 1, inplace = True)
```

### `Fare`

```python
# Divide Fare into four bands

titanic_df['FareBand'] = pd.cut(titanic_df['Fare'], 4)
titanic_df[['FareBand', 'Survived']].groupby(['FareBand'], as_index=False).mean().sort_values(by = 'FareBand')
```

There are 4 ranges of Fare:
- 0 : (-0.003, 0.75]
- 1 : (0.75, 1.5]
- 2 : (1.5, 2.25]
- 3 : (2.25, 3.0]

```python
# Assign ordinal to each fare band
titanic_df.loc[titanic_df['Fare'] <= 0.75, 'Fare'] = 0
titanic_df.loc[(titanic_df['Fare'] > 0.75) & (titanic_df['Fare'] <= 1.5), 'Fare'] = 1
titanic_df.loc[(titanic_df['Fare'] > 1.5) & (titanic_df['Fare'] <= 2.25), 'Fare'] = 2
titanic_df.loc[titanic_df['Fare'] > 2.25, 'Fare'] = 3
```

```python
# Convert Fare into integer

titanic_df['Fare'] = titanic_df['Fare'].astype('int')
# Drop FareBand feature

titanic_df = titanic_df.drop('FareBand', axis = 1)
titanic_df.head()
```

### `Age`

```python
# Divide Age into five bands

titanic_df['AgeBand'] = pd.cut(titanic_df['Age'], 5)
titanic_df[['AgeBand', 'Survived']].groupby(['AgeBand'], as_index=False).mean().sort_values(by = 'AgeBand')
```

There are 5 ranges of Age:
- 0 : (0.346, 15.136]
- 1 : (15.136, 29.852]
- 2 : (29.852, 44.568]
- 3 : (44.568, 59.284]
- 4 : (59.284, 74.0]

```python
# Assign ordinals to each age band 

titanic_df.loc[titanic_df['Age'] <= 15.136, 'Age'] = 0
titanic_df.loc[(titanic_df['Age'] > 15.136) & (titanic_df['Age'] <= 29.852), 'Age'] = 1
titanic_df.loc[(titanic_df['Age'] > 29.852) & (titanic_df['Age'] <= 44.568), 'Age'] = 2
titanic_df.loc[(titanic_df['Age'] > 44.568) & (titanic_df['Age'] <= 59.284), 'Age'] = 3
titanic_df.loc[titanic_df['Age'] > 59.284 , 'Age'] = 4
# Drop age band feature

titanic_df.drop('AgeBand', axis = 1, inplace = True)
```

```python
# Convert ordinal Age into integer

titanic_df['Age'] = titanic_df['Age'].astype('int')
```

## 5. Modelling

Scikit-learn is one of the most popular libraries for machine learning in Python and that is what we will use in the modelling part of this project.

Since Titanic is a classfication problem, we will need to use classfication models, also known as classifiers, to train on our model to make predictions.I have chosen the following classifiers for the job:
- Logistic Regression
- K-nearest neighbours

In this section of the notebook, I will fit the models to the training set and evaluate their accuracy at making predictions. Once the best model is determined, I will also do hyperparameter tuning to further boost the performance of the best model.

#### 5.1 Split the dataset into trainning and testing set

We need to first split our data into independent variables or predictor variables, represented by X as well as dependent variable or response variable, represented by y.

y_train is the survived column in our training set and X_train is the other columns in the training set excluding the Survived column. Our models will learn to classify survival, y_train based on all X_train and make predictions on X_test.

```python
train, test = train_test_split(titanic_df, test_size = 0.1, random_state = 142 )
X_train = train.drop(['Survived'], axis = 1)
y_train = train['Survived']
X_test = test.drop(['Survived'], axis = 1)
y_test = test['Survived']

print(train.shape, test.shape)
```

#### 5.2 Logistic Regression

```python
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
```

#### Use First Model To Predict The Test Data

```python
log_preds = log_reg.predict(X_test)
```

#### Evaluate The Logistic Model

```python
log_accuracy = round(accuracy_score(y_test,log_preds)*100, 2)
print(f' The accuracy of Logistic Regression model : {log_accuracy}')
```

#### 5.3 KNN Model

```python
clf = KNeighborsClassifier(n_neighbors = 10)
clf.fit(X_train, y_train)
```

#### Predict the Test Dataset

```python
knn_preds = clf.predict(X_test)
```

#### Evaluate The KNN Model

```python
knn_accuracy = round(accuracy_score(y_test,knn_preds)*100, 2)
print(f' The accuracy of Logistic Regression model : {knn_accuracy}')
```

From the results, the KNN model provide the best performance. Therefore, I will do the hyper-parameter tuning for KNN model.

## 6. Tune the hyper-parameter K in KNN

#### Create A KNN Classifier Model And Use K-Fold Cross Validation For Testing Accuracy Estimation

```python
clf_k_cross = KNeighborsClassifier(n_neighbors = 10)
score = cross_val_score(clf_k_cross, X_train, y_train, cv = 5 )
print('Accuracy for each round: \n', score)
print('Average accuracy: %.4f +- %.4f' % (score.mean(), score.std()))
```

#### Plot The Accurcy Trend

```python
train_error = []
test_error = []

for i in range(1, 101):
    clf = KNeighborsClassifier(n_neighbors = i)
    clf.fit(X_train, y_train)
    train_error.append(1-accuracy_score(clf.predict(X_train), y_train))
    test_error.append(1- accuracy_score(clf.predict(X_test), y_test))

# Plot the diagram
plt.plot(range(1,101), train_error, marker = 'x', label = 'Train Error')
plt.plot(range(1,101), test_error, marker = 'o', label = 'Test Error')
plt.xlabel('K')
plt.ylabel('Error')
plt.legend(loc='best')
plt.show()
```

#### Tuning the hyper-parameter Using GridSearchCV

```python
# Define the search space for parameters
parameter_grid = {'n_neighbors': range(1,51)}

# Create the machine learning model:
knn_clf = KNeighborsClassifier()
grid_search = GridSearchCV(knn_clf, parameter_grid, scoring = 'accuracy', cv = 5)
grid_results = grid_search.fit(X_train, y_train)

# Identify the best parameter
print(f"Best K value: {grid_results.best_params_['n_neighbors']}")
print(f'Best accuracy: {round(grid_results.best_score_*100,2)}')
```

#### Train data with new hyper-parameter from GridSearchCV method

```python
Grid = KNeighborsClassifier(n_neighbors = grid_results.best_params_['n_neighbors'])
Grid.fit(X_train, y_train)
```

#### Predict the test dataset and evaluation

```python
Grid_predict = Grid.predict(X_test)
acc_Grid = round(accuracy_score(y_test,Grid_predict)*100, 2)
print(f'The accuracy of KNN model with new-hyper parameter from GridSearchCV method : {acc_Grid}')
```

#### Plot the results from 3 models

```python
# Create a dataframe to visualise the performance
performance_df = pd.DataFrame({
    'Model' : ['LogisticRegression', 'KNN', 'KNN using GridSearchCV'],
    'Accuracy' : [log_accuracy, knn_accuracy, acc_Grid]
})

# Plot the figure
# plt.figure(figsize = (5, 5))
sns.set(style = 'whitegrid')
ax = sns.barplot(x=performance_df['Model'], y=performance_df['Accuracy'])
for container in ax.containers:
    ax.bar_label(container)
ax.set_ylim([70, 100]);
```
